**Lendo os dados**

```python
#Importando biblioteca para le os dados
import pandas as pd
```

```python
#lendo arquivo csv
amostra_df = pd.read_csv("/content/amostra-para-exercicio.csv")
```

```python
#Lendo dados
amostra_df.head()
```

```python
#Verificando dados duplicados 
amostra_df[amostra_df.duplicated('qtd_de_matriculas')]
```

```python
#Excluindo dados duplicados e colocando em uma nova variavel
df = amostra_df.drop_duplicates()
```

**Ultilizando sql**

```python
#Instalando biblioteca pandassql
pip install pandasql
```

```python
#Importando biblioteca
import pandasql as ps
```

```python
#Usando sql para filtrar dados da coluna nome_curso distintos
query = """Select Distinct nome_curso from df"""
ps.sqldf(query, locals())
```

```python
#Utilizando sql para deletar valores iguais a 0 da coluna qtd_de_matricula
query = """Delete from df where qtd_de_matriculas=0 """
ps.sqldf(query.lower(), locals())
```

```python
#Importando biblioteca para o grafico
import  seaborn as sns
import matplotlib.pyplot as plt


```

**Criando Grafico boxplot**

```python
#Grafico boxplot utilizando as colunas nome_escola, qtd_de_matricula
plt.figure(figsize=(10,10))
sns.boxplot(x = "nome_escola", y = "qtd_de_matriculas", data=df)
```
